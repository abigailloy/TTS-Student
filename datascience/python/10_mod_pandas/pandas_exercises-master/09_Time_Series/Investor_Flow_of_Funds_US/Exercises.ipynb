### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv). 

### Step 3. Assign it to a variable called 

```python
#assigns dataset to variable 'url'
url = 'https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv'

#creates dataframe 'specialPaper' from url
specialPaper = pd.read_csv(url)

#prints first 5 elements
specialPaper.head()
```

### Step 4.  What is the frequency of the dataset?

```python
#prints all info about dataset, frequency for the date included
specialPaper.describe(include='all')
#weekly data
```

### Step 5. Set the column Date as the index.

```python
#sets Date column as index
specialPaper = specialPaper.set_index('Date')
specialPaper.head()
```

### Step 6. What is the type of the index?

```python
#gets index info, including type
specialPaper.index
```

### Step 7. Set the index to a DatetimeIndex type

```python
#sets index to Datetime index type
specialPaper.index = pd.to_datetime(specialPaper.index)
type(specialPaper.index)
```

### Step 8.  Change the frequency to monthly, sum the values and assign it to monthly.

```python
#changes frequency to monthly, sums values, and assigns to variable 'monthly'
monthly = specialPaper.resample('M').sum()
monthly
```

### Step 9. You will notice that it filled the dataFrame with months that don't have any data with NaN. Let's drop these rows.

```python
#brutally takes out the undesirables 
monthly = monthly.dropna()
monthly
```

### Step 10. Good, now we have the monthly data. Now change the frequency to year.

```python
years = monthly.resample('AS-JAN').sum()
years
```
